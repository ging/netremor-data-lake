```python
# Lista todos los archivos en el directorio de Subjects
subject_directory = "/home/datalake/data/Subjects"
subject_ids = [name for name in os.listdir(subject_directory) if os.path.isdir(os.path.join(subject_directory, name))]
print(subject_ids)
# Inicializar una lista para almacenar los DataFrames
dataframes = []
# Iterar sobre cada subject ID
for subject_id in subject_ids:
    try:
        # Construir la ruta a la tabla Delta del sujeto
        delta_table_path = os.path.join(subject_directory, subject_id, "Continuo")
        # Verificar si la ruta existe y es un directorio
        if os.path.exists(delta_table_path) and os.path.isdir(delta_table_path):
            df = DeltaTable(f"/home/datalake/data/Subjects/{subject_id}/Continuo").to_pandas()
            dataframes.append(df)
    except Exception as e:
        print(f"Error al procesar el subject ID {subject_id}: {e}")

# Concatenar todos los DataFrames en uno solo
final_df = pd.concat(dataframes, ignore_index=True)
# Mostrar el DataFrame con todos los records continuos
final_df.head(10)
```

```python
continuous_record_ids = final_df['record_id']
file_paths = []

# Iterar sobre cada subject ID
for subject_id in subject_ids:
    # Construir la ruta al directorio Tasks del sujeto
    task_directory = os.path.join(subject_directory, subject_id, "Tasks")
    
    # Verificar si el directorio Tasks existe
    if os.path.exists(task_directory) and os.path.isdir(task_directory):
        # Iterar sobre cada continuous_record_id
        for record_id in continuous_record_ids:
            # Construir la ruta al archivo usando subject_id y record_id
            file_path = os.path.join(task_directory, str(record_id))
            
            # Verificar si el archivo existe
            if os.path.exists(file_path):
                df = DeltaTable(f"/home/datalake/data/Subjects/{subject_id}/Tasks/{record_id}").to_pandas()
                file_paths.append(df)
            else:
                print(f"El archivo {file_path} no existe para el subject ID {subject_id}.")
    else:
        print(f"El directorio {task_directory} no existe o no es un directorio válido para el subject ID {subject_id}.")

# Mostrar las rutas de los archivos encontrados
continuos_df = pd.concat(file_paths, ignore_index=True)
continuos_df.describe()
```

```python
task_id_counts = continuos_df['task_id'].value_counts()

plt.figure(figsize=(10, 6))
task_id_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Tarea realizada')
plt.ylabel('Frecuencia')
plt.title('Tareas realizadas por los sujetos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
continuos_df['starts_at'] = pd.to_numeric(continuos_df['starts_at'])
continuos_df['ends_at'] = pd.to_numeric(continuos_df['ends_at'])

continuos_df['medida'] = continuos_df['ends_at'] - continuos_df['starts_at']


maximo = continuos_df.loc[continuos_df['medida'].idxmax()]
minimo = continuos_df.loc[continuos_df['medida'].idxmin()]



# Convertir las diferencias a segundos
maximo_segundos = maximo['medida'] / 1000
minimo_segundos = minimo['medida'] / 1000

# Imprimir los resultados
print("Entrada con la medición más larga:")
print(maximo)
print(f"Diferencia en segundos: {maximo_segundos} segundos")
print("\nEntrada con la medición más corta:")
print(minimo)
print(f"Diferencia en segundos: {minimo_segundos} segundos")
```
